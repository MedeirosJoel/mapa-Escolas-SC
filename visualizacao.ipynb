```python
import geopandas as pd
import folium
from folium import plugins
```

```python
geodf = pd.read_file('dataframe.csv', encoding='utf-8')
geodf['geometry'] = pd.points_from_xy(geodf['lng'], geodf['lat'])
```

```python
bounds = [[-10, -20], [10, 20]]

mapa = folium.Map()
heat_map = folium.plugins.HeatMap(geodf[['lat', 'lng']], radius=15)
heat_map.add_to(mapa)
mapa
```
